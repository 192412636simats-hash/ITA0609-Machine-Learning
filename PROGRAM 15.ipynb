```python
# Weather Data Analyzer

temps = []

for i in range(7):
    t = float(input("Enter Temperature for Day " + str(i + 1) + ": "))
    temps.append(t)

print("Maximum Temperature:", max(temps))
print("Minimum Temperature:", min(temps))
print("Average Temperature:", sum(temps) / len(temps))
```
