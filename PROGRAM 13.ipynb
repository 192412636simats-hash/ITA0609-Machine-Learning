```python
# Parking Fee Calculator

hours = int(input("Enter Parking Hours: "))
vehicle = input("Enter Vehicle Type (Bike/Car): ")

if vehicle.lower() == "bike":
    fee = hours * 20
else:
    fee = hours * 50

print("Parking Fee = ₹", fee)
```
