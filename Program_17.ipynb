```python
# Program 17: Password Strength Checker

# Read password
password = input("Enter Password: ")

# Initialize variables
upper = 0
lower = 0
digit = 0
special = 0

# Check each character
for ch in password:
    if ch.isupper():
            upper = 1
    if ch.islower():
            lower = 1
    if ch.isdigit():
            digit = 1
    if not ch.isalnum():
            special = 1
  # Check password strength
    if len(password) >= 8 and upper == 1 and lower == 1 and digit == 1 and special == 1:
           print("Strong Password")

    if len(password) < 8 or upper == 0 or lower == 0 or digit == 0 or special == 0:
           print("Weak Password")
```
