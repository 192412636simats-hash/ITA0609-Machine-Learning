```python
# Program 20: Movie Ticket Booking System

# Read booking details
seats = int(input("Enter Number of Seats: "))
category = input("Enter Seat Category (Silver/Gold/Platinum): ")

# Set ticket price
price = 0

if category.lower() == "silver":
    price = 150
if category.lower() == "gold":
    price = 250
if category.lower() == "platinum":
    price = 350

 # Calculate total cost
total = seats * price

 # Initialize discount
discount = 0

# Apply discount
if seats >= 5:
  discount = total * 0.10

 # Calculate final amount
final_amount = total - discount

 # Display bill
print("Total Cost =", total)
print("Discount =", discount)
print("Final Amount =", final_amount)
```
