```python
# Mobile Recharge Billing

amount = float(input("Enter Recharge Amount: "))

if amount >= 500:
    cashback = 50
else:
    cashback = 20

final = amount - cashback

print("Cashback:", cashback)
print("Final Amount:", final)
```
