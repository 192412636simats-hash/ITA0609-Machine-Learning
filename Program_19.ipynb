```python
# Program 19: Inventory Stock Management

# Read stock details
stock = int(input("Enter Current Stock: "))
minimum = int(input("Enter Minimum Stock Level: "))

# Check stock status
if stock < minimum:
    print("Stock is Low! Please Restock.")
if stock >= minimum:
    print("Stock Level is Sufficient.")
```
