```python
# Railway Ticket Fare Calculator

fare = float(input("Enter Base Fare: "))
age = int(input("Enter Age: "))

if age < 12:
    fare = fare * 0.5
elif age >= 60:
    fare = fare * 0.7

print("Ticket Fare = ₹", fare)
```
