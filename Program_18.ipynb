```python
# Program 18: Blood Donation Eligibility Checker

# Read age, weight and hemoglobin
age = int(input("Enter Age: "))
weight = float(input("Enter Weight (kg): "))
hemoglobin = float(input("Enter Hemoglobin Level: "))

# Check eligibility
if age >= 18 and age <= 65 and weight >= 50 and hemoglobin >= 12.5:
    print("Eligible for Blood Donation")
if age < 18 or age > 65 or weight < 50 or hemoglobin < 12.5:
    print("Not Eligible for Blood Donation")
```
